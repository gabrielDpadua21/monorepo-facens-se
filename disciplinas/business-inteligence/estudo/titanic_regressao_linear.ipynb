# Estudo de regressão linear usando base do Titanic

## Modelo usando regressão lienar simples

```r
titanic_train <- read.csv(file = '../data/titanic-train.csv')
```

```r
titanic_test <- read.csv(file = '../data/titanic-test.csv')
```

```r
head(titanic_train)
```

```r
head(titanic_test)
```

### Modelo usando hodoult e regressão linear simples

```r
modelo <- lm(Survived ~ Sex, titanic_train)
summary(modelo)
```

```r
y_predict <- predict(modelo, titanic_test)
```

```r
y_real <- titanic_train$Survived
```

```r
mse <- mean((y_real - y_predict)^2)
```

```r
mse
```

### Modelo usando holdout e regressão linear multipla

```r
modelo <- lm(Survived ~ Sex + Age + Pclass, titanic_train)
summary(modelo)
```

```r
y_predict <- predict(modelo, titanic_test)
```

```r
y_real <- titanic_train$Survived
```

```r
mse < mean((y_real - y_predict)^2)
```

```r
mse
```

## Modelo usando amostragem aleatória e bootstrap com regressão linear multipla

```r
max <- 1000
set.seed(123)

mse <- vector(length = max)

for(i in 1:max) {
    indexTrain = sample(1:nrow(titanic_train), 500, TRUE)
    indexTest = sample(1:nrow(titanic_test), 200, TRUE)
    
    train <- titanic_train[indexTrain, ]
    test <- titanic_test[indexTest, ]
    
    modelo <- lm(Survived ~ Sex + Age + Pclass + Embarked, train)
    y_predict <- predict(modelo, test[, c('Sex', 'Age', 'Pclass', 'Embarked')])
    y_real <- titanic_train$Survived
    
    mse[i] <- mean((y_real - y_predict)^2)
}
```

```r
mse
```
