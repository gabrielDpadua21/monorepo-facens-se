### Separando a base de dados em treino e teste

#### Exemplo de sorteio de valores

#### Sorteando linhas da base de dados boston

```r
index <- sample(1:nrow(Boston), 0.7*nrow(Boston))
```

#### Gerando a base de dados de treinamento

```r
train <- Boston[index, ]
```

#### Gerando a base de dados de teste

```r
test <- Boston[-index, ]
```

#### Criando o modelo com regressão linear

```r
modelo <- lm(crim ~ ., train)
summary(modelo)
```

#### Realizando uma previsão

```r
y_predict = predict(modelo, test[,-1])
```

```r
y_real = test$crim
```

### Verificando as métricas

### MAE

```r
mean(abs(y_real - y_predict))
```

### MSE

```r
mean((y_real - y_predict)^2)
```

### RMSE

```r
sqrt(mean((y_real - y_predict)^2))
```
