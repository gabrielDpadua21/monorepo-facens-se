### Modelo C

```r
modelo <- lm(Sales ~ Price + Urban + US, Carseats)
summary(modelo)
```

1) No modelo aplicado como p-value é menor que 5% então a regressão existe e é significativa

2) As variaveis que explicam as vendas (Sales) são Price (sendo preço da empresa em cada local) e US (sendo uma variavel qualitativa indicando se a loja está localizada nos Estados Unidos ou não).

3) Como R quadrado ajustado é de 0.2335 a relação entre as 3 variáveis é baixa

### Interpretação

Para cada dolar a mais no preço serão vendidos menos 54 unidade

### Equação matemática do modelo

Sales = 13.043469 - 0.054459 * Price - 0.021916 * Urban (Yes) + 1.200573 * US (yes)
